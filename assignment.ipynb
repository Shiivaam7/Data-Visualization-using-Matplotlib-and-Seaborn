```python
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
years = list(range(2011, 2021))
kohli_runs = [1000, 1200, 900, 1100, 1300, 1400, 1500, 1600, 1450, 1550]
rohit_runs = [800, 850, 950, 1200, 1250, 1350, 1400, 1380, 1420, 1500]
sehwag_runs = [1100, 1000, 980, 1050, 1020, 950, 900, 880, 870, 850]
```

```python
plt.plot(years, kohli_runs, color='blue', linestyle='-', marker='o', label='Kohli')
plt.plot(years, rohit_runs, color='green', linestyle='--', marker='s', label='Rohit sharma')
plt.plot(years, sehwag_runs, color='red' , linestyle= '-.', marker='^',label='Sehwag')
plt.xlabel('years')
plt.ylabel('runs')
plt.title('Runs Comparison over 10 years')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
