```python
import matplotlib.pyplot as plt
plt.style.use("ggplot") #choose any style you like.
```

```python
labels =["sachin" , "sehwag" , "kholi" , "yuvraj"]
runs = [18000, 8000 , 12000 , 9500]
plt.title = ("carrer runs of indian batsman")
plt.pie(runs , labels=labels)
plt.show()
```

```python
plt.pie(
    runs,
    labels = labels,
    wedgeprops={'edgecolor':'black' ,'linewidth':2 , 'linestyle': '--'}

)
plt.show()
```

```python
explode = [0.1, 0, 0, 0]  # Only highlight Sachin's slice

plt.pie(runs, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    shadow=True  # adds a 3D-like shadow
    
)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    shadow=True,
    startangle=140
)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140
)
plt.show()
```
